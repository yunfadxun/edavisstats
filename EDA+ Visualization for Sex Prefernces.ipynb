```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("sex_pref.csv")
```

```python
df.info()
```

```python
for i in df.select_dtypes(include=["bool","object"]):
    plt.figure()
    sns.countplot(df[i], hue="Gender", data=df)
    plt.title(f"countplot of {i} breakdown by Gender")
    plt.show()
```

```python
for i in df.select_dtypes(include=["bool","object"]):
    plt.figure()
    plt.pie(df[i].value_counts(), labels=df[i].unique(), autopct= "%1.1f%%")
    plt.title(f"pie chart for {i}")
```

```python
plt.figure()
sns.countplot(y= "Gender",data=df, orient="h")
```

```python
plt.figure()
sns.countplot("Favorite Color",data=df, order=df["Favorite Color"].value_counts(ascending=False).index)
plt.show()
```

```python
from scipy.stats import chi2_contingency

cross_tab={}
for i in df.columns[0:4]:
    cross_tab[i]=pd.crosstab(df[i],df["Gender"])

for column_name, crosstab_df in cross_tab.items():
    print(column_name)
    print(crosstab_df)
    chi2,p,_,_= chi2_contingency(crosstab_df)
    print(f"x2 value for this column against gender is: {chi2: .3f}")
    print(f"p-value is: {p: .3f}")
    print()
```
